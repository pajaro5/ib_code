```python
import random

#Write your function here:
def check_card(card):
    # TODO: Implement using NESTED IF statements
    # Check if card is EVEN and MULTIPLE OF 3
    # Print "You win!" if both conditions are true
    if card % 2 == 0:
        if card % 3 ==0:
            print("You win!")


def main():
    # Create the deck
    deck = [1, 2, 3, 4, 5, 6, 7, 8, 9]

    # Shuffle the deck
    random.shuffle(deck)
    
    # Draw one card (remove from deck)
    drawn_card = deck.pop(0)
    
    # Print the drawn card
    print(f"Card drawn: {drawn_card}")
    
    # Check the card using your function
    check_card(drawn_card)

# Run the game
if __name__ == "__main__":
    main()

```
