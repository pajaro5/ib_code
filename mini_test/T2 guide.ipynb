## 1️⃣ Manual Logic: Counting and Summing

### 📖 Concept
To replace `len()` and `sum()`, we use a loop. 
* **Counter:** Starts at 0, adds 1 every time the loop runs.
* **Accumulator:** Starts at 0, adds the **value** of the current element.

```python
# DEMONSTRATION: Recreating len() and sum()
data = [10, 20, 30, 40]

count = 0
total = 0

for number in data:
    count += 1    # Manual len()
    total += number # Manual sum()

print(f"Elements: {count}")
print(f"Total Sum: {total}")
```

## 2️⃣ Manual Logic: Finding Extremes (The Max/Min Logic)

### 📖 Concept: Identifying Peaks and Valleys
In programming, finding an "extreme" value means identifying the **highest** (Maximum) or **lowest** (Minimum) element in a collection. Since the IB prohibits the use of `max()` or `min()`, we must use a single loop to scan the data.

**The "Champion Method":**
1.  **The Baseline:** We define a variable (the "Champion") and initialize it with the first element of the list (`lst[0]`). At the start, this number is technically both the largest and smallest we have seen.
2.  **The Challenge:** We iterate through the list. Every time we find a new number, we compare it to our current "Champion".
3.  **The Crown:** * For **Maximum**: If the new number is *greater* than our champion, the new number takes the crown.
    * For **Minimum**: If the new number is *smaller* than our champion, it becomes the new lowest value.



### 💻 Demonstration: Manual Max and Min in a Single Pass
Observe how we find both extremes without altering the list or using multiple loops.


```python
def find_extremes(lst):
    # Step 1: Initialize champions with the first item
    # We assume lst[0] is the current winner for both categories
    highest_peak = lst[0]
    lowest_valley = lst[0]
    
    # Step 2: Scan the list (The Challenge)
    for number in lst:
        
        # Step 3: Check for a new Maximum
        if number > highest_peak:
            highest_peak = number # Update the maximum champion
            
        # Step 4: Check for a new Minimum
        if number < lowest_valley:
            lowest_valley = number # Update the minimum champion
            
    return highest_peak, lowest_valley

# Usage
results = [45, 12, 89, 3, 50]
max_val, min_val = find_extremes(results)
print(f"System Maximum: {max_val} | System Minimum: {min_val}")
```

## 3️⃣ Sorting: The Bubble Sort

Bubble Sort is a logical process used to organize a list by comparing **adjacent pairs**. 

* **The Mechanism:** The algorithm looks at two neighbors. If the left one is larger than the right one, they **swap** positions.
* **The "Bubble":** In each full pass through the list, the largest unsorted value "bubbles up" like an air bubble in water until it reaches its correct position at the end.
* **The Security Rule (`n - 1`):** We must stop the inner loop one step before the end. Why? Because we compare the current item (`j`) with the next one (`j + 1`). If we go to the very last item, there is no "next neighbor," and the program will crash.


```python
# DEMONSTRATION: Ascending Bubble Sort
def bubble_sort(arr):
    number_of_elements = 0
    for i in arr: 
        number_of_elements += 1 # Manual len()
    
    for i in range(number_of_elements):
        for j in range(0, number_of_elements - i-1):
            if arr[j] > arr[j + 1]:
                # Swap logic using a temporary variable
                temp = arr[j]
                arr[j] = arr[j + 1]
                arr[j + 1] = temp
    return arr

print("Sorted:", bubble_sort([5, 1, 4, 2]))
```

## 4️⃣ OOP: Encapsulation and Statics

### 📖 Concept
* **Constructor (`__init__`):** Sets up the object when it's born.
* **Private (`__`):** Hides data for safety (Encapsulation).
* **Static Attribute:** Shared by ALL objects of the class (e.g., a counter for total objects created).
* **Instance Attribute:** Unique to each object (e.g., a serial number).

```python
# DEMONSTRATION: Static vs Instance & Private attributes
class Car:
    total_cars = 0  # STATIC: Shared by all cars

    def __init__(self, model, price):
        self.model = model      # INSTANCE: Unique to this car
        self.__price = price    # PRIVATE: Encapsulated (cannot access directly)
        Car.total_cars += 1     # Updating static variable

    def get_price(self):
        return self.__price

c1 = Car("Tesla", 50000)
c2 = Car("Ford", 30000)

print(f"Created: {Car.total_cars} cars.")
print(f"Tesla Price: {c1.get_price()}")
```

## 6️⃣ Database Design & Data Integrity (Topic A3)

### 📖 Concept: Attributes and Data Types
In a database, an **Entity** (like a Weather Reading) is described by **Attributes**. Each attribute must have a specific **Data Type** so the system knows how to store and process the information correctly.

* **Integer:** Whole numbers (e.g., Wind Speed in km/h).
* **Float:** Numbers with decimals (e.g., Temperature 22.5).
* **String/Varchar:** Text (e.g., City Name or Weather Description).
* **Date/Timestamp:** Accurate recording of dates and times.

### 📖 Concept: Primary Keys & Auto-increment
A **Primary Key (PK)** is a field that **uniquely identifies** every record in a table. No two rows can ever have the same PK.

**Why use Auto-increment?**
Instead of a human typing an ID (which leads to mistakes or duplicate numbers), the database automatically generates the next number (1, 2, 3...). This ensures **Data Integrity** because it guarantees uniqueness without human error.

### 📖 Concept: Relational Databases vs. Flat-files
A **Flat-file** is a simple text file (like `.txt` or `.csv`) where data is just a list. A **Relational Database** uses structured, linked tables.

**Advantages of Relational Databases for long-term data:**
* **Speed:** Much faster to search through years of data using indexes.
* **No Redundancy:** Avoids saving the same information multiple times.
* **Relationships:** Can easily connect "Weather Readings" with "Station Locations" in different tables.
