### ¬øQu√© es un bucle `for`?
Un bucle es una estructura de control que repite un bloque de instrucciones un n√∫mero predeterminado de veces. En Python, la forma m√°s eficiente de repetir un c√≥digo un n√∫mero espec√≠fico de veces es utilizando la funci√≥n `range()` [4].

```python
# El range(5) genera los n√∫meros del 0 al 4.
# La variable 'i' toma cada uno de esos valores en cada iteraci√≥n.
for i in range(5):
    print("Iteraci√≥n n√∫mero:", i)
```

### Controlando el `range()`
La funci√≥n `range()` puede recibir hasta tres argumentos para darte control total: `range(inicio, fin, paso)`.
* **Inicio:** Desde qu√© n√∫mero empezamos (incluido).
* **Fin:** Hasta qu√© n√∫mero llegamos (excluido).
* **Paso:** De cu√°nto en cu√°nto avanzamos.

```python
print("N√∫meros pares del 2 al 10:")
for numero in range(2, 11, 2):
    print(numero)

print("\nCuenta regresiva:")
for numero in range(5, 0, -1): # Usamos un paso negativo para retroceder
    print(numero)
```

### Iterando sobre Cadenas de Texto (Strings)
Adem√°s de n√∫meros, el bucle `for` puede recorrer otras estructuras, como las cadenas de caracteres [5].

```python
palabra = "PYTHON"
for letra in palabra:
    print("Letra:", letra)
```

### Contadores y Acumuladores
Podemos usar variables fuera del bucle para llevar la cuenta de lo que sucede adentro.
* **Contador:** Suma un valor fijo (ej. +1) en cada iteraci√≥n.
* **Acumulador:** Suma un valor variable en cada iteraci√≥n para calcular un total.

```python
suma_total = 0 # Esto es un acumulador
contador_vueltas = 0 # Esto es un contador

for i in range(1, 6):
    suma_total = suma_total + i
    contador_vueltas = contador_vueltas + 1
    print(f"Vuelta {contador_vueltas}: sumando {i}, total acumulado = {suma_total}")
```

### Variables de control descartables (`_`)
A veces necesitamos repetir una acci√≥n, pero no nos importa el valor num√©rico de la iteraci√≥n. Por convenci√≥n, los programadores de Python usan un guion bajo (`_`) como variable por defecto cuando no van a utilizarla dentro del bucle [6].

```python
for _ in range(3):
    print("¬°No me importa el n√∫mero de iteraci√≥n, solo quiero saludar!")
```

### Bucles Anidados (Un bucle dentro de otro)
Un bucle anidado se crea colocando un bucle (el bucle interno) dentro de otro (el bucle externo) [7]. Por cada iteraci√≥n del bucle externo, el bucle interno se ejecuta por completo antes de que el bucle externo pase a su siguiente ciclo [7].

```python
# Simularemos 3 horas, y por cada hora, solo 3 "minutos" para no hacer muy largo el output
for hora in range(1, 4):
    for minuto in range(1, 4):
        print(f"Reloj: {hora} horas y {minuto} minutos")
    print("--- Termina una hora completa ---")
```

### Generando Tablas de Multiplicar
Los bucles anidados son una forma excelente de generar combinaciones, como las tablas de multiplicar [8].

```python
# Imprimir las tablas del 1 y del 2
for numero_tabla in range(1, 3):
    print(f"=== Tabla del {numero_tabla} ===")
    for multiplicador in range(1, 11):
        resultado = numero_tabla * multiplicador
        print(f"{numero_tabla} x {multiplicador} = {resultado}")
```
