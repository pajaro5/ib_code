### ¿Qué es un bucle `while`?
A diferencia del bucle `for` (que itera una cantidad de veces definida), el bucle `while` **se repite mientras una condición específica se mantenga verdadera** [4]. Es útil cuando no sabemos de antemano el número exacto de iteraciones.

**¡Cuidado!** Es tu responsabilidad declarar una variable de control antes del bucle y *actualizarla* dentro de él. Si lo olvidas, crearás un bucle infinito que nunca terminará [1].

```python
contador = 1
while contador <= 5:
    print("Iteración número:", contador)
    contador = contador + 1  # ¡Crucial! Actualizamos la variable para que el bucle termine.
```

### Acumuladores con `while`
Al igual que con el `for`, podemos acumular valores en una variable sumando o multiplicando números ciclo tras ciclo hasta que la condición límite se alcance.

```python
numero = 1
suma_total = 0

while numero <= 10:
    suma_total = suma_total + numero
    numero = numero + 1

print(f"La suma de los números del 1 al 10 es: {suma_total}")
```

### Control forzado: La instrucción `break`
A veces necesitamos salir de un bucle inmediatamente, sin importar si la condición principal del `while` sigue siendo verdadera. Para eso usamos la instrucción `break` [2].

```python
limite = 1

while limite <= 10:
    if limite == 6:
        print("¡Límite de emergencia alcanzado! Abortando.")
        break  # Rompe el bucle por completo inmediatamente.
    print("Todo bajo control. Nivel:", limite)
    limite = limite + 1
```

### Saltando un paso: La instrucción `continue`
Si quieres omitir el código que queda en la iteración actual y saltar directamente a evaluar de nuevo la condición para la siguiente vuelta, usas `continue` [2].

```python
paso = 0

while paso < 5:
    paso = paso + 1
    if paso == 3:
        print("-> ¡Me salto el número 3!")
        continue  # Omite lo que queda debajo y vuelve al while.
    print("Paso actual:", paso)
```

### El poder del `while`: Iteraciones desconocidas
El bucle `while` es indispensable cuando **no sabemos con certeza cuántas veces** se debe repetir un proceso [4]. Por ejemplo, reducir matemáticamente un número iterativamente.

```python
valor = 100
vueltas = 0

# Seguiremos dividiendo entre 2 mientras el valor sea mayor a 1
while valor > 1:
    valor = valor / 2
    vueltas = vueltas + 1
    print(f"Vuelta {vueltas}: el valor se redujo a {valor}")
```

### Bucles `while` Anidados
Al igual que con el `for`, podemos anidar bucles `while`. La clave de la gestión anidada aquí es recordar que **debes reiniciar manualmente** el contador del bucle interno antes de que empiece a dar sus propias vueltas.

```python
fila = 1
while fila <= 3:
    columna = 1  # Este reinicio es OBLIGATORIO en cada vuelta de la fila
    while columna <= 3:
        print(f"Coordenada ({fila}, {columna})")
        columna = columna + 1
    print("--- Termina la fila ---")
    fila = fila + 1
```
